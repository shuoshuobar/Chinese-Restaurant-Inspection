```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.colors as mcolors
import textwrap
from matplotlib.colors import LinearSegmentedColormap
import matplotlib as mpl

# --- Ensure SVG output keeps text elements (not paths)
mpl.rcParams['svg.fonttype'] = 'none'

# Data
data = {
    'Violation_Description': [
        "Rodents/Insects/Pests Issue", 
        "Foods Temperature Control", 
        "Improper Pest/Pesticide Management", 
        "Unprotected from Contamination Source", 
        "Cleaning Accessibility", 
        "Improper Draining/Sewage Disposal", 
        "Improper/Contaminated Food Contact Surface", 
        "Adulterated/Contaminated Food", 
        "Deficient Hand Washing Facility", 
        "Improper Thawing Procedure"
    ],
    'Count': [184, 76, 52, 41, 36, 28, 22, 21, 18, 12]
}

df = pd.DataFrame(data)
df = df.sort_values('Count', ascending=False)

# Normalize counts for color mapping
max_count = df['Count'].max()
min_count = df['Count'].min()
df['Normalized_Count'] = (df['Count'] - min_count) / (max_count - min_count)

# Color gradient from red (high) to blue (low)
cmap = LinearSegmentedColormap.from_list("RedToBlue", ["#E41A1C", "#984EA3", "#377EB8"])

# Map violations to colors
color_map = {
    row['Violation_Description']: mcolors.rgb2hex(cmap(1 - row['Normalized_Count']))
    for _, row in df.iterrows()
}

# Layout helper
def enhanced_treemap_layout(sizes):
    total = sum(sizes)
    sizes = [s/total for s in sizes]

    def divide_area(items, x, y, width, height):
        if not items:
            return []
        if len(items) == 1:
            return [{'x': x, 'y': y, 'width': width, 'height': height, 'index': items[0][0]}]
        if width > height:
            return divide_horizontally(items, x, y, width, height)
        else:
            return divide_vertically(items, x, y, width, height)

    def divide_horizontally(items, x, y, width, height):
        result = []
        total_size = sum(item[1] for item in items)
        cumulative, cut_index = 0, 0
        for i, (idx, size) in enumerate(items):
            cumulative += size
            if cumulative >= total_size / 2:
                cut_index = i
                break
        left_items = items[:cut_index+1]
        right_items = items[cut_index+1:]
        left_total = sum(item[1] for item in left_items)
        left_width = (left_total / total_size) * width
        if left_items:
            result.extend(divide_area(left_items, x, y, left_width, height))
        if right_items:
            result.extend(divide_area(right_items, x + left_width, y, width - left_width, height))
        return result

    def divide_vertically(items, x, y, width, height):
        result = []
        total_size = sum(item[1] for item in items)
        cumulative, cut_index = 0, 0
        for i, (idx, size) in enumerate(items):
            cumulative += size
            if cumulative >= total_size / 2:
                cut_index = i
                break
        top_items = items[:cut_index+1]
        bottom_items = items[cut_index+1:]
        top_total = sum(item[1] for item in top_items)
        top_height = (top_total / total_size) * height
        if top_items:
            result.extend(divide_area(top_items, x, y, width, top_height))
        if bottom_items:
            result.extend(divide_area(bottom_items, x, y + top_height, width, height - top_height))
        return result

    enumerated_sizes = list(enumerate(sizes))
    rects = divide_area(enumerated_sizes, 0, 0, 1, 1)
    rects.sort(key=lambda r: r['index'])
    for rect in rects:
        del rect['index']
    return rects

layout = enhanced_treemap_layout(df['Count'].values)

# Text wrapper
def wrap_text(text, rect_width, rect_height):
    aspect_ratio = rect_width / rect_height
    char_width = 0.01
    chars_per_line = int(rect_width / char_width)
    chars_per_line = max(10, min(chars_per_line, 40))
    if aspect_ratio > 2:
        chars_per_line = int(chars_per_line * 1.5)
    if aspect_ratio < 0.5:
        chars_per_line = int(chars_per_line * 0.7)
    return textwrap.fill(text, chars_per_line, break_long_words=False)

# Plot
fig, ax = plt.subplots(figsize=(12, 8))

for i, (rect, (_, row)) in enumerate(zip(layout, df.iterrows())):
    desc = row['Violation_Description']
    color = color_map[desc]
    r = patches.Rectangle(
        (rect['x'], rect['y']), rect['width'], rect['height'],
        linewidth=2, edgecolor='white', facecolor=color
    )
    ax.add_patch(r)
    wrapped = wrap_text(desc, rect['width'], rect['height'])
    area = rect['width'] * rect['height']
    fontsize = max(8, min(np.sqrt(area) * 40, 14))
    if rect['width'] > 0.05 and rect['height'] > 0.03:
        ax.text(
            rect['x'] + rect['width']/2, rect['y'] + rect['height']/2,
            wrapped,
            color='white', fontsize=fontsize, fontweight='bold',
            fontfamily='Arial', ha='center', va='center', wrap=True
        )

ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis('off')
fig.patch.set_facecolor('white')

# Color legend
gradient = np.linspace(0, 1, 256)
gradient = np.vstack((gradient, gradient))
ax_legend = fig.add_axes([0.25, 0.05, 0.5, 0.02])
ax_legend.imshow(gradient, aspect='auto', cmap=cmap)
ax_legend.set_title('Violation Occurrence', fontsize=10, pad=5)
ax_legend.set_xticks([0, 255])
ax_legend.set_xticklabels(['Highest', 'Lowest'])
ax_legend.set_yticks([])

plt.tight_layout(pad=0)
plt.subplots_adjust(left=0, right=1, top=0.95, bottom=0.15)

# Save SVG with real text
plt.savefig('violation_treemap.svg', format='svg', bbox_inches='tight')

# Show
plt.show()
```

```python
import pandas as pd
import numpy as np
import textwrap
import matplotlib.colors as mcolors
from matplotlib.colors import LinearSegmentedColormap
import xml.etree.ElementTree as ET
import xml.dom.minidom as minidom

# === Data Setup ===
data = {
    'Violation_Description': [
        "Rodents/Insects/Pests Issue", 
        "Foods Temperature Control", 
        "Improper Pest/Pesticide Management", 
        "Unprotected from Contamination Source", 
        "Cleaning Accessibility", 
        "Improper Draining/Sewage Disposal", 
        "Improper/Contaminated Food Contact Surface", 
        "Adulterated/Contaminated Food", 
        "Deficient Hand Washing Facility", 
        "Improper Thawing Procedure"
    ],
    'Count': [184, 76, 52, 41, 36, 28, 22, 21, 18, 12]
}

df = pd.DataFrame(data).sort_values('Count', ascending=False)

# Normalize for color mapping
df['Normalized_Count'] = (df['Count'] - df['Count'].min()) / (df['Count'].max() - df['Count'].min())

# Color gradient (Red to Blue)
cmap = LinearSegmentedColormap.from_list("RedToBlue", ["#E41A1C", "#984EA3", "#377EB8"])
color_map = {
    row['Violation_Description']: mcolors.rgb2hex(cmap(1 - row['Normalized_Count']))
    for _, row in df.iterrows()
}

# === Treemap Layout Function ===
def enhanced_treemap_layout(sizes):
    total = sum(sizes)
    sizes = [s / total for s in sizes]

    def divide_area(items, x, y, width, height):
        if len(items) == 1:
            return [{'x': x, 'y': y, 'width': width, 'height': height, 'index': items[0][0]}]
        return divide_horizontally(items, x, y, width, height) if width > height else divide_vertically(items, x, y, width, height)

    def divide_horizontally(items, x, y, width, height):
        total_size = sum(s for _, s in items)
        cut_index, cum = 0, 0
        for i, (_, s) in enumerate(items):
            cum += s
            if cum >= total_size / 2:
                cut_index = i
                break
        left, right = items[:cut_index+1], items[cut_index+1:]
        left_width = sum(s for _, s in left) / total_size * width
        return divide_area(left, x, y, left_width, height) + divide_area(right, x + left_width, y, width - left_width, height)

    def divide_vertically(items, x, y, width, height):
        total_size = sum(s for _, s in items)
        cut_index, cum = 0, 0
        for i, (_, s) in enumerate(items):
            cum += s
            if cum >= total_size / 2:
                cut_index = i
                break
        top, bottom = items[:cut_index+1], items[cut_index+1:]
        top_height = sum(s for _, s in top) / total_size * height
        return divide_area(top, x, y, width, top_height) + divide_area(bottom, x, y + top_height, width, height - top_height)

    items = list(enumerate(sizes))
    layout = divide_area(items, 0, 0, 1, 1)
    layout.sort(key=lambda d: d['index'])
    for r in layout:
        r.pop('index')
    return layout

layout = enhanced_treemap_layout(df['Count'].values)

# === Improved Word Wrapping Function ===
def improved_text_wrap(text, rect_width, rect_height):
    """
    Better word wrapping that breaks only at word boundaries
    and adjusts for larger fonts
    """
    # Use fewer chars per line to allow for larger font
    chars_per_line = max(3, int(rect_width * 10))
    
    # If very small rect, just use first word
    if rect_width < 0.08 or rect_height < 0.08:
        words = text.split()
        return words[0] if words else ""
    
    # Break at word boundaries only
    words = text.split()
    lines = []
    current_line = []
    current_length = 0
    
    for word in words:
        word_length = len(word)
        
        # Add word to current line or start new line
        if current_length + word_length + 1 <= chars_per_line or not current_line:
            current_line.append(word)
            current_length += word_length + 1
        else:
            lines.append(" ".join(current_line))
            current_line = [word]
            current_length = word_length + 1
    
    # Add the last line
    if current_line:
        lines.append(" ".join(current_line))
    
    # Limit lines based on available height (less lines = bigger font)
    max_lines = max(1, int(rect_height * 10))
    if len(lines) > max_lines:
        lines = lines[:max_lines]
    
    return '\n'.join(lines)

# === Function for determining text color based on background ===
def get_text_color(bg_color):
    # Convert hex to RGB
    bg_color = bg_color.lstrip('#')
    r, g, b = int(bg_color[0:2], 16)/255, int(bg_color[2:4], 16)/255, int(bg_color[4:6], 16)/255
    
    # Calculate luminance - standard formula
    luminance = 0.299 * r + 0.587 * g + 0.114 * b
    
    # Return white for dark backgrounds, black for light backgrounds
    return 'white' if luminance < 0.5 else 'black'

# === Direct SVG generation for Illustrator compatibility ===
# SVG dimensions
svg_width = 1200
svg_height = 800
margin = 50

# Create SVG root
root = ET.Element('svg', {
    'xmlns': 'http://www.w3.org/2000/svg',
    'width': str(svg_width),
    'height': str(svg_height),
    'viewBox': f'0 0 {svg_width} {svg_height}'
})

# Add title
title = ET.SubElement(root, 'title')
title.text = 'Food Safety Violations Treemap'

# Create a defs section for gradient
defs = ET.SubElement(root, 'defs')

# Add a style sheet that ensures text is selectable in Illustrator
style = ET.SubElement(defs, 'style', {'type': 'text/css'})
style.text = """
    text {
        font-family: Arial, sans-serif;
        -webkit-user-select: text;
        -moz-user-select: text;
        -ms-user-select: text;
        user-select: text;
        cursor: text;
    }
    .rect-label {
        dominant-baseline: middle;
        text-anchor: middle;
        pointer-events: none;
    }
"""

# Create a group for all elements
main_group = ET.SubElement(root, 'g', {'transform': f'translate({margin},{margin})'})

# Canvas dimensions (adjusted for margins)
canvas_width = svg_width - 2 * margin
canvas_height = svg_height - 2 * margin

# Add rectangles and text
for i, (rect, (_, row)) in enumerate(zip(layout, df.iterrows())):
    # Get data
    desc = row['Violation_Description']
    color = color_map[desc]
    count = row['Count']
    
    # Calculate actual coordinates
    x = rect['x'] * canvas_width
    y = rect['y'] * canvas_height
    width = rect['width'] * canvas_width
    height = rect['height'] * canvas_height
    
    # Create rectangle group
    rect_group = ET.SubElement(main_group, 'g', {'id': f'rect-group-{i}'})
    
    # Add rectangle
    rectangle = ET.SubElement(rect_group, 'rect', {
        'x': str(x),
        'y': str(y),
        'width': str(width),
        'height': str(height),
        'fill': color,
        'stroke': 'white',
        'stroke-width': '2'
    })
    
    # Add the violation count as an attribute for later reference
    rectangle.set('data-count', str(count))
    
    # Use improved word wrapping function
    processed_text = improved_text_wrap(desc, rect['width'], rect['height'])
    
    # Only add text if there's content
    if processed_text:
        # Get appropriate text color
        text_color = get_text_color(color)
        
        # Count lines to adjust font size
        lines = processed_text.split('\n')
        
        # LARGER FONT SIZES based on rectangle dimensions and line count
        if len(lines) <= 1:
            # Single line - largest font
            font_size = min(32, max(14, min(width, height) * 0.22))
        elif len(lines) == 2:
            # Two lines - medium font
            font_size = min(28, max(12, min(width, height) * 0.18))
        else:
            # Multiple lines - smaller font
            font_size = min(24, max(10, min(width, height) * 0.14))
        
        # Calculate text position
        text_x = x + width / 2
        text_y = y + height / 2
        
        # Calculate line height based on font size (tighter spacing)
        line_height = font_size * 1.1
        
        # Adjust vertical position for multi-line text
        y_offset = (len(lines) - 1) * line_height / 2
        
        # Create text group
        text_group = ET.SubElement(rect_group, 'text', {
            'x': str(text_x),
            'y': str(text_y - y_offset),
            'font-size': str(font_size),
            'font-weight': 'bold',
            'fill': text_color,
            'class': 'rect-label',
            'id': f'text-{i}'
        })
        
        # Add each line of text as a tspan
        for j, line in enumerate(lines):
            tspan = ET.SubElement(text_group, 'tspan', {
                'x': str(text_x),
                'dy': '0' if j == 0 else str(line_height),
                'text-anchor': 'middle'
            })
            tspan.text = line

# Add gradient legend
legend_width = canvas_width * 0.5
legend_height = 20
legend_x = (canvas_width - legend_width) / 2
legend_y = canvas_height + 20

# Create legend group
legend_group = ET.SubElement(main_group, 'g', {'id': 'legend'})

# Gradient definition
gradient = ET.SubElement(defs, 'linearGradient', {
    'id': 'legendGradient',
    'x1': '0%',
    'y1': '0%',
    'x2': '100%',
    'y2': '0%'
})

# Gradient stops
stops = [
    ('0%', '#E41A1C'),
    ('50%', '#984EA3'),
    ('100%', '#377EB8')
]

for offset, color in stops:
    ET.SubElement(gradient, 'stop', {
        'offset': offset,
        'stop-color': color
    })

# Legend rectangle
ET.SubElement(legend_group, 'rect', {
    'x': str(legend_x),
    'y': str(legend_y),
    'width': str(legend_width),
    'height': str(legend_height),
    'fill': 'url(#legendGradient)'
})

# Legend title - LARGER FONT
legend_title = ET.SubElement(legend_group, 'text', {
    'x': str(legend_x + legend_width / 2),
    'y': str(legend_y - 10),
    'font-size': '16',
    'text-anchor': 'middle'
})
legend_title.text = 'Violation Occurrence'

# Legend labels - LARGER FONT
ET.SubElement(legend_group, 'text', {
    'x': str(legend_x),
    'y': str(legend_y + legend_height + 15),
    'font-size': '14',
    'text-anchor': 'start'
}).text = 'Highest'

ET.SubElement(legend_group, 'text', {
    'x': str(legend_x + legend_width),
    'y': str(legend_y + legend_height + 15),
    'font-size': '14',
    'text-anchor': 'end'
}).text = 'Lowest'

# Convert to string with proper formatting
rough_string = ET.tostring(root, 'utf-8')
reparsed = minidom.parseString(rough_string)
pretty_svg = reparsed.toprettyxml(indent="  ")

# Write to file
with open('violation_treemap_illustrator.svg', 'w', encoding='utf-8') as f:
    f.write(pretty_svg)

print("SVG created successfully with larger selectable text optimized for Illustrator")
```
