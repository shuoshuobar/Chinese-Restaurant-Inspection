```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.colors as mcolors
from matplotlib.colors import LinearSegmentedColormap
import textwrap
```

```python
# Create the dataset with original violation descriptions
data = {
    'Violation_Description': [
        "Foods Temperature Control", 
        "Rodents/Insects/Pests Harborage/Conditions",
        "Mice/Rats",
        "Not Protected from Contamination Source",
        "Roaches",
        "Cleaning Accessibility",
        "No Pest Management Professional Contract",
        "Improper Draining/Sewage Disposal",
        "Improper/Contaminated Food Contact Surface",
        "Flies",
        "Adulterated/Contaminated Food",
        "Deficient Hand Washing Facility",
        "Pesticide Misuse/Improper Storage",
        "Improper Thawing Procedure",
        "Inadequate Personal Cleanliness",
        "Missing Food Protection Certificate",
        "Improper Wiping Cloths Storage/Sanitization",
        "Deficient Dish/Ware Washing",
        "Improper Garbage Receptacle Maintenance",
        "Improper Single-service Article Handling"
    ],
    'Count': [76, 67, 57, 41, 39, 36, 29, 28, 22, 21, 21, 18, 13, 12, 11, 10, 9, 5, 5, 5]
}

# Create DataFrame
df = pd.DataFrame(data)

# Sort by count in descending order
df = df.sort_values('Count', ascending=False)

# Normalize counts for color mapping (0 to 1 scale)
max_count = df['Count'].max()
min_count = df['Count'].min()
df['Normalized_Count'] = (df['Count'] - min_count) / (max_count - min_count)

# Create a color gradient from red (highest) to blue (lowest)
cmap = LinearSegmentedColormap.from_list("RedToBlue", ["#E41A1C", "#984EA3", "#377EB8"])

# Map each violation to a color based on its count
color_map = {}
for _, row in df.iterrows():
    # Get color from the colormap based on normalized count
    color = cmap(1 - row['Normalized_Count'])
    # Convert RGBA to hex
    hex_color = mcolors.rgb2hex(color)
    color_map[row['Violation_Description']] = hex_color

# Create figure and axes
fig, ax = plt.subplots(figsize=(12, 8))

# Enhanced treemap layout algorithm
def enhanced_treemap_layout(sizes):
    """Calculate rectangle positions for treemap for all items."""
    # Normalize sizes
    total = sum(sizes)
    sizes = [s/total for s in sizes]
    
    # Function to divide space
    def divide_area(items, x, y, width, height):
        if not items:
            return []
        
        # For a single item, use the entire area
        if len(items) == 1:
            return [{'x': x, 'y': y, 'width': width, 'height': height, 'index': items[0][0]}]
        
        # Decide whether to divide horizontally or vertically
        if width > height:
            return divide_horizontally(items, x, y, width, height)
        else:
            return divide_vertically(items, x, y, width, height)
    
    # Divide horizontally
    def divide_horizontally(items, x, y, width, height):
        result = []
        total_size = sum(item[1] for item in items)
        
        # Find mid point based on size
        cumulative = 0
        cut_index = 0
        
        for i, (idx, size) in enumerate(items):
            cumulative += size
            if cumulative >= total_size / 2:
                cut_index = i
                break
        
        # Left group
        left_items = items[:cut_index+1]
        left_total = sum(item[1] for item in left_items)
        left_width = (left_total / total_size) * width
        
        # Right group
        right_items = items[cut_index+1:]
        
        # Recursively process each group
        if left_items:
            result.extend(divide_area(left_items, x, y, left_width, height))
        
        if right_items:
            result.extend(divide_area(right_items, x + left_width, y, width - left_width, height))
        
        return result
    
    # Divide vertically
    def divide_vertically(items, x, y, width, height):
        result = []
        total_size = sum(item[1] for item in items)
        
        # Find mid point based on size
        cumulative = 0
        cut_index = 0
        
        for i, (idx, size) in enumerate(items):
            cumulative += size
            if cumulative >= total_size / 2:
                cut_index = i
                break
        
        # Top group
        top_items = items[:cut_index+1]
        top_total = sum(item[1] for item in top_items)
        top_height = (top_total / total_size) * height
        
        # Bottom group
        bottom_items = items[cut_index+1:]
        
        # Recursively process each group
        if top_items:
            result.extend(divide_area(top_items, x, y, width, top_height))
        
        if bottom_items:
            result.extend(divide_area(bottom_items, x, y + top_height, width, height - top_height))
        
        return result
    
    # Create rectangles
    enumerated_sizes = list(enumerate(sizes))
    rects = divide_area(enumerated_sizes, 0, 0, 1, 1)
    
    # Sort rects by index
    rects.sort(key=lambda r: r['index'])
    
    # Remove index from output
    for rect in rects:
        del rect['index']
    
    return rects

# Calculate layout for all violations
layout = enhanced_treemap_layout(df['Count'].values)

# Function to wrap text with appropriate line length based on rectangle size
# Ensuring line breaks only happen after complete words
def wrap_text(text, rect_width, rect_height):
    # Estimate how many characters can fit horizontally
    aspect_ratio = rect_width / rect_height
    char_width = 0.01  # Approximate width of a character relative to plot width
    
    # Calculate how many characters can fit in the width
    chars_per_line = int(rect_width / char_width)
    
    # Set a minimum and maximum number of characters per line
    chars_per_line = max(10, min(chars_per_line, 40))
    
    # For very wide rectangles, use more chars per line
    if aspect_ratio > 2:
        chars_per_line = int(chars_per_line * 1.5)
    
    # For very tall rectangles, use fewer chars per line
    if aspect_ratio < 0.5:
        chars_per_line = int(chars_per_line * 0.7)
    
    # Wrap the text using textwrap.fill which ensures breaks only at word boundaries
    # Setting break_long_words=False ensures that words longer than width are not broken
    wrapped_text = textwrap.fill(text, chars_per_line, break_long_words=False, break_on_hyphens=True)
    
    return wrapped_text

# Draw the rectangles
for i, (rect, (_, row)) in enumerate(zip(layout, df.iterrows())):
    description = row['Violation_Description']
    color = color_map[description]
    
    # Create rectangle
    r = patches.Rectangle(
        (rect['x'], rect['y']), rect['width'], rect['height'],
        linewidth=2, edgecolor='white', facecolor=color,
        alpha=1.0
    )
    ax.add_patch(r)
    
    # Format the text with proper wrapping
    wrapped_text = wrap_text(description, rect['width'], rect['height'])
    
    # Calculate the appropriate font size based on rectangle size
    area = rect['width'] * rect['height']
    fontsize = np.sqrt(area) * 40
    fontsize = max(8, min(fontsize, 14))  # Clamp between 8 and 14
    
    # Only add text if rectangle is large enough
    if rect['width'] > 0.05 and rect['height'] > 0.03:
        ax.text(
            rect['x'] + rect['width']/2, rect['y'] + rect['height']/2,
            wrapped_text,
            color='white', fontsize=fontsize, fontweight='bold', 
            fontfamily='Arial', horizontalalignment='center',
            verticalalignment='center', wrap=True
        )

# Set limits and remove axes
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis('off')

# Set background color to white
fig.patch.set_facecolor('white')

# Add title
# plt.title('Most health code violations indicate improper temperature control for of hot and cold foods', 
        # fontsize=16, fontweight='bold', fontfamily='Arial', pad=20)

# Create a color gradient for the legend
gradient = np.linspace(0, 1, 256)
gradient = np.vstack((gradient, gradient))

# Add a color bar legend
ax_legend = fig.add_axes([0.25, 0.05, 0.5, 0.02])
ax_legend.imshow(gradient, aspect='auto', cmap=cmap)
ax_legend.set_title('Violation Occurrence', fontsize=10, pad=5)
ax_legend.set_xticks([0, 255])
ax_legend.set_xticklabels(['Highest', 'Lowest'])
ax_legend.set_yticks([])

# Avoid "auto" padding
plt.tight_layout(pad=0)
plt.subplots_adjust(left=0, right=1, top=0.95, bottom=0.15)

plt.savefig('violation_treemap.svg', format='svg', bbox_inches='tight')

# Show the plot
plt.show()
```
